```python
meas = df['Measurements'].apply(lambda cell: pd.Series(json.loads(cell.replace("'", '"'))))
meas
```

```python
hero = pd.concat([df, meas], axis = 1)
hero
```

```python
hero=hero.replace({'Height':'[a-z]'},'',regex=True)
hero['Height']=hero['Height'].apply(lambda row: float(row))
hero=hero.replace({'Weight':'[a-z]'},'',regex=True)
hero['Weight']=hero['Weight'].apply(lambda row: float(row))
hero.info()
```

```python
hero.loc[2,'Powers']
```

```python
hero['NewPowers'] = hero['Powers'].str.replace('"',"'")
```

```python
hero.loc[2,'NewPowers']
```

```python
exploded = hero.explode('NewPowers')
exploded[['NewPowers']].head()
```

```python
cols_to_make = exploded['NewPowers'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df[col] = df['NewPowers'].str.contains(col)
df.head()
```
